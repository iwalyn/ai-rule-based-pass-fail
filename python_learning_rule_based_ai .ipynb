```python
import pandas as pd

df = pd.read_csv("python_learning_exam_performance.csv")

df.head(10)
```

```python
df.shape
```

```python
def rule_based_pass_fail(row):
    if (
        row["hours_spent_learning_per_week"] >= 5 and
        row["practice_problems_solved"] >= 30 and
        row["projects_completed"] >= 2
    ):
        return 1  # Passs
    else:
        return 0  # Fail
```

```python
df["predicted_passed_exam"] = df.apply(rule_based_pass_fail, axis=1)

df[["passed_exam", "predicted_passed_exam"]].head(10)
```

```python
accuracy = (df["passed_exam"] == df["predicted_passed_exam"]).mean()
accuracy
# Accuracy: Doru tahmin edilen renci orann ifade eder.
```

```python
error_rate = 1 - accuracy
error_rate
# Error Rate: Yanl tahmin edilen renci orann ifade eder.
```
